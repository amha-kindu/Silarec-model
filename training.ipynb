```python
import tensorflow as tf
import numpy as np

tf.random.set_seed(42)
```

```python
X = np.load('x_359_batch.npy')
y = np.load('y_359_batch.npy')
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3423)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import os
from tensorflow.keras.callbacks import TensorBoard

log_dir = os.path.join('performance_log')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input


model = Sequential([
    Input(shape=(50, 258)),
    LSTM(64, return_sequences=True, activation='relu'),
    LSTM(128, return_sequences=True, activation='relu'),
    LSTM(64, return_sequences=False, activation='relu'),
    Dense(64, activation='relu'),
    Dense(64, activation='relu'),
    Dense(359, activation='softmax')
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.00001),
    loss = tf.keras.losses.categorical_crossentropy,
    metrics=tf.keras.metrics.categorical_accuracy
)
```

```python
epoch = 2000
training_history = model.fit(
    X_train, 
    y_train, 
    epochs=epoch, 
    callbacks=[
        tb_callback,
    ])
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['figure.figsize'] = (18, 8)
rcParams['axes.spines.top'] = False
rcParams['axes.spines.right'] = False 

plt.plot(
    np.arange(1, epoch+1), 
    training_history.history['loss'], 
    label='Loss', lw=3
)

plt.plot(
    np.arange(1
              , epoch+1), 
    training_history.history['categorical_accuracy'], 
    label='Accuracy', lw=3
    
)

plt.title('Evaluation metrics(learning rate: 1e-05)', size=20)
plt.xlabel('Epoch', size=14)
plt.legend()

# plt.savefig(r'slarec_model\v1.0\evaluation_metrics.jpg')
plt.show()
```
